```python
import cv2
import os
import glob
import matplotlib.pyplot as plt

def load_images(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
graphs = glob.glob("/home/vamsimocherla/Pictures/Screenshot*")
```

```python
fig=plt.figure(figsize=(40, 20))
columns = 3
rows = 2
for i, graph in enumerate(graphs):
    img = cv2.imread(graph)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(img)

path = os.path.sep.join(["/home/vamsimocherla/PycharmProjects/dl4od", "adam_vs_sgd.png"])
plt.savefig(path)
plt.show()
```
